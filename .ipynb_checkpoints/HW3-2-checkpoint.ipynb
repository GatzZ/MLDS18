```python
image_ids = []
eyes = []
hairs = []

with open('AnimeDataset/tags_clean.csv') as f:
    for l in f:
        if 'eyes' in l and 'hair' in l:
            print(l, l.split(',')[0])
            image_ids.append(l.split(',')[0])
            es, hs = [], []
            for t in l.split('\t'):                
                if 'eyes' in t: 
                    es.append(t.split(':')[0])
                if 'hair' in t: 
                    hs.append(t.split(':')[0])
            eyes.append(es)
            hairs.append(hs)
```

```python
hair_cates = ['orange hair', 'white hair', 'aqua hair', 'gray hair',
 'green hair', 'red hair', 'purple hair', 'pink hair',
  'blue hair', 'black hair', 'brown hair', 'blonde hair']

eyes_cates = ['black eyes', 'orange eyes',
 'pink eyes', 'yellow eyes', 'aqua eyes', 'purple eyes',
 'green eyes', 'brown eyes', 'red eyes', 'blue eyes']

all_cates = hair_cates + eyes_cates
```

```python
CODE_LEN = len(all_cates)

def onehot_enc(hairs, eyes):
    assert len(hairs) == len(eyes)
    code_ls = []
    
    for hs, es in zip(hairs, eyes):
        code = [0] * CODE_LEN
        for h in hs:
            if h in all_cates:
                idx = all_cates.index(h)
                code[idx] = 1
        for e in es:
            if e in all_cates:
                idx = all_cates.index(e)
                code[idx] = 1
        code_ls.append(np.array(code))
    return code_ls

onehot_codes = onehot_enc(hairs, eyes)
```

```python
image_files = [x + '.jpg' for x in image_ids]
df = pd.DataFrame({'image_files': image_files, 'codes': onehot_codes})
df.loc[:, 'index'] = df.index.tolist()
code_ser = df['codes']

image_datagen = ImageDataGenerator(preprocessing_function=lambda x: (x / 127.5) -1)
image_generator = image_datagen.flow_from_dataframe(
        df, 
        'AnimeDataset/faces/',
        x_col = 'image_files',
        y_col = 'index',
        target_size=(64, 64),
        batch_size=4,
        class_mode='other',
        seed=1)


def get_train_generator(image_generator):
    for imgs, idxes in image_generator:
        yield imgs, np.vstack(code_ser.iloc[idxes].values)

train_generator = get_train_generator(image_generator)
imgs, codes = next(train_generator)
```
